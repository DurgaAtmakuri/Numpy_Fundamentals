```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Universities.csv')

```

```python
df
```

```python
#plot histogram for sat column
plt.hist(df['SAT'])
plt.xlabel('SAT')
plt.ylabel('Count')
plt.show() #to avoid the values in the output
```

```python
#plot histogram for sat column
plt.figure(figsize=(5,5))
plt.hist(df['SAT'])
plt.xlabel('SAT')
plt.ylabel('Count')
plt.title('hist of SAT')
plt.show() #to avoid the values in the output
```

```python
sns.histplot(df['SAT'])
plt.xlabel('SAT')
plt.ylabel('Count')
plt.title('hist of SAT')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.histplot(df['SAT'], color='green', edgecolor='black')
plt.xlabel('SAT')
plt.ylabel('Count')
plt.title('hist of SAT')
plt.show()
```

```python
plt.plot(df['SAT'], df['Accept'], 'rs') #for blue 'bo and red 'rs'
plt.xlabel('SAT')
plt.ylabel('Accept')
plt.title('Scatter plot between SAT vs ACCEPT')
plt.show()
```

```python
# prompt: create a column between top 10 column and accept column and use blue colour circle markers in the scatter plot.

import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is already loaded as in your provided code
# ... (your existing code to load the dataframe) ...

# Create the scatter plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.scatter(df['SAT'], df['Accept'], marker='o', s=50, c='blue', label='Data Points')


# Add labels and title
plt.xlabel('SAT Scores')
plt.ylabel('Acceptance Rate')
plt.title('Scatter Plot of SAT Scores vs. Acceptance Rate')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
